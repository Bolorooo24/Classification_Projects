## which one should you use? Pyplot vs Matplotlib 00 method?

* when plotting something quickly, okay to use the pyplot
* when plotting sth advanced, use 00 method


```python
heart_desease.head(10)
```

```python
over_50 = heart_desease[heart_desease["age"]>50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter", 
            x = "age",
            y = "chol",
            c = "target",
            figsize=(10,6))
```

```python
# 00method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
            x= "age",
            y="chol",
            c="target",
            ax =ax)
##ax.set_xlim([45,100])
```

## OO method from scratch

```python
## OO method
fig, ax = plt.subplots(figsize = (10,6))

## plot the data
scatter= ax.scatter(x=over_50["age"],
                   y = over_50["chol"],
                   c = over_50["target"]);

##customize

ax.set(title="Hear disease and Cholerstral leverls",
      xlabel = "Age",
      ylabel = "Cholesterol");

## Add a legent
ax.legend(*scatter.legend_elements(), title = "Target");

## Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle ="--" );
```

```python
over_50
```

```python
#subplot of chol, age, 
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols =1,
                              figsize = (10,10),
                              sharex= True);

#Add dta to ax0
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# Customize ax0

ax0.set(title = "Heart Disease and Cholesterol Levels",
      
       ylabel = "Cholestoral");

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle = "--")

# add date to ax1
scatter = ax1.scatter(x = over_50["age"],
                     y=over_50["thalach"],
                     c= over_50["target"])
# Customize ax1, labeling
ax1.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Thalach")

# add a legend to ax1 
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle= "--");
# add a title
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold");
```

## customizing the y and x axis limitations

```python
#subplot of chol, age, 
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols =1,
                              figsize = (10,10),
                              sharex= True);

#Add dta to ax0
scatter = ax0.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "winter")

# Customize ax0

ax0.set(title = "Heart Disease and Cholesterol Levels",
      
       ylabel = "Cholestoral");
# change the x axis limits
ax0.set_xlim([50, 80])

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle = "--")

# add date to ax1
scatter = ax1.scatter(x = over_50["age"],
                     y=over_50["thalach"],
                     c= over_50["target"],
                     cmap = "winter")
# Customize ax1, labeling
ax1.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Thalach")

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])


# add a legend to ax1 
ax1.legend(*scatter.legend_elements(), title="Target")

#add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle= "--");
# add a title
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold");
```
